```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import os
import pandas as pd
```

```python
labels = ['自転車', 'バス', '車']
sizes = [25, 40, 35]

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1,1,1)

# 円グラフの描画
ax.pie(sizes, labels=labels)

plt.show()
```

```python
# 円グラフの書式を設定するサンプル
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1, 1, 1)

# 扇形の書式設定用dictionary
wedge_props = {'edgecolor': 'black', 'linewidth': 2}

# テキストの書式設定用dictionary
text_props = {'fontsize': 18}

ax.pie(sizes,
      explode=(0., 0.05, 0),
      labels=labels,
      autopct='%1.0f%%',
      pctdistance=0.5,
      shadow=False,
      labeldistance=1.35,
      startangle=90,
      radius=0.3,
      counterclock=False,
      wedgeprops=wedge_props,
      textprops=text_props,
      center=(0.5, 0.5),
      frame=True)

plt.show()
```

```python
df = pd.read_csv('../sample_data/anime_genre_top10_pivoted.csv', index_col='genre')
df.head
```

```python
tv_df = df.sort_values(by='TV', ascending=False)['TV']

tv_df
```

```python
movie_df = df.sort_values(by='Movie', ascending=False)['Movie']

movie_df
```

```python
fig = plt.figure(figsize=(9,4))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# カラーセット
colors1 = ('gold', 'coral', 'plum', 'orchid', 'lightseagreen',
          'yellowgreen', 'lightskyblue', 'pink', 'cornflowerblue',
          'orangered')
colors2 = ('coral', 'orangered', 'plum', 'pink', 'gold',
          'cornflowerblue', 'yellowgreen', 'lightseagreen', 'orchid',
          'lightskyblue')


ax1.pie(tv_df,
       explode=(0, 0, 0, 0, 0, 0, 0.15, 0, 0, 0.15),
       labels=tv_df.index,
       autopct='%1.0f%%',
       colors=colors1,
       startangle=90,
       counterclock=False)

ax2.pie(movie_df,
       explode=(0, 0, 0, 0, 0, 0, 0.15, 0, 0, 0.15),
       labels=movie_df.index,
       autopct='%1.0f%%',
       colors=colors2,
       startangle=90,
       counterclock=False)
        
ax1.set_title('TV')
ax2.set_title('Movie')

fig.subplots_adjust(wspace=0.3) # サブプロット間のスペース調整

plt.show()
```
