```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
plt.style.use('ggplot')
```

```python
# データセットの作成
mu = 100 # 平均値
sigma = 10 # 標準偏差
np.random.seed(0) # sampleようなので乱数固定
x =np.random.normal(mu, sigma, 10000)

# ヒストグラムの描画
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(x)

plt.show()
```

```python
fig2 = plt.figure()
ax = fig2.add_subplot(1, 1, 1)
ax.hist(x, rwidth=0.9, bins=16)

plt.show()
```

```python
df = pd.read_csv('../sample_data/anime_master.csv', index_col='anime_id')
df.head()
```

```python
fig2 = plt.figure()
ax2  = fig2.add_subplot(1, 1, 1)

ax2.hist(df['rating'], range(0, 10), rwidth=0.9)
ax2.set_title('Rating')

plt.show()
```

```python
fig3 = plt.figure()
ax3 = fig3.add_subplot(1, 1, 1)
df_tv = df[df['type'] == 'TV']

ax3.hist(df_tv['episodes'], rwidth=0.9)
ax3.set_title('Episodes')

plt.show()
```

```python
fig4 = plt.figure()
ax4 = fig4.add_subplot(1, 1, 1)
df_tv = df[df['type'] == 'TV']

ax4.hist(df_tv['episodes'], rwidth=0.9, range=(0,100))
ax4.set_title('Episodes(0-100)')

plt.show()
```

```python
# 横向きのヒストグラムの描画
np.random.seed(0)
x = np.random.normal(100, 10, 10000)

fig5 = plt.figure()
ax5  = fig5.add_subplot(1, 1, 1)

# 横向きを指定する
ax5.hist(x, rwidth=0.9, bins=16, orientation='horizontal')

plt.show()
```

```python
# 相対度数のヒストグラムの描画
fig6 = plt.figure()
ax6  = fig6.add_subplot(1, 1, 1)

# normedをTrueに指定
ax6.hist(df['rating'], normed=True, rwidth=0.9)

plt.show()
```

```python
# 累積ヒストグラムのsample
fig7 = plt.figure()
ax7  = fig7.add_subplot(1, 1, 1)

# cumulativeをTrueに指定
ax7.hist(df['rating'], normed=True, rwidth=0.9, cumulative=True)

plt.show()
```

```python
# 階級幅を設定したヒストグラムの描画
fig8 = plt.figure()
ax8  = fig8.add_subplot(1, 1, 1)

ax8.hist(df['rating'], bins=[2, 4, 5.5, 6.5, 7, 7.5, 8.5, 10], rwidth=0.9)
plt.show()
```

```python
# 近似曲線を追加する例
bins = 50
dfmin = np.min(df['rating'])
dfmax = np.max(df['rating'])

fig9 = plt.figure()
ax9  = fig9.add_subplot(1, 1, 1)

# ヒストグラムの描画
ax9.hist(df['rating'], bins=bins, range=(dfmin, dfmax), normed=True, rwidth=0.9)

# 平均と標準偏差
mean,sigma = df['rating'].mean(), df['rating'].std()

# X value
x = np.linspace(dfmin, dfmax, bins) # ビンの区切りの位置
# 近似的な確率密度関数を使用してY値を生成
y = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(- (x -mean)**2 / (2 * sigma**2))

# 近似曲線の描画
ax9.plot(x,y)

plt.show()
```

```python
# 複数グループのヒストグラムを重ねて描画
types = df['type'].unique()
labels = types.tolist()

fig10 = plt.figure(figsize=(8,6))
ax10  = fig10.add_subplot(1, 1, 1)

b_num = np.arange(0, 10.5, 0.5)

for t in types:
  ax10.hist(df.loc[df['type'] == t, 'rating'],
    bins=b_num,
    rwidth = 0.9,
    alpha = 0.5,
    label=t
  )

ax10.legend()
ax10.set_xlabel('rating')
ax10.set_ylabel('Count(rating)')

plt.show()
```

```python
# 複数グループを並べて描画
dataset = [df.loc[df['type'] == t, 'rating'] for t in types]

fig11 = plt.figure(figsize=(8, 6))
ax11  = fig11.add_subplot(1, 1, 1)

ax11.hist(dataset,
  bins=np.arange(0, 10.5, 0.5),
  rwidth=0.9,
  alpha=0.8,
  label=labels
)

ax11.legend()
ax11.set_xlabel('rating')
ax11.set_ylabel('Count(rating)')

plt.show()
```

```python
# 積み上げヒストグラムの描画
# dataset, labelsは、「複数グループを並べて描画」で作成したものを使用する
fig12 = plt.figure()
ax12  = fig12.add_subplot(1, 1, 1)

ax12.hist(
    dataset,
    bins=np.arange(0, 10.5, 0.5),
    rwidth=0.9,
    alpha=0.7,
    label=labels,
    stacked=True
)

ax12.legend()
ax12.set_xlabel=('rating')
ax12.set_ylabel('Count(rating)')

plt.show()
```
