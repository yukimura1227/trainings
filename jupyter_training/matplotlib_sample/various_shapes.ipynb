# Matplotlibで色々な図形を描画するサンプル

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

## 円弧を描画する

```python
# Arc円弧
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1,1,1)

c1 = mpatches.Arc(xy=(0.4,0.5), width=0.4, height=0.4, angle=30, theta1=0, theta2=230,linewidth=10, color='lightblue')
ax.add_patch(c1)

c2 = mpatches.Arc(xy=(0.6,0.5), width=0.4, height=0.4, angle=30, theta1=0, theta2=230,linewidth=10, color='pink')
ax.add_patch(c2)

plt.show
```

## 矢印を描画する

```python
# 矢印を描画する
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1,1,1)

a1 = mpatches.Arrow(x=0.1, y=0.4, dx=0.6, dy=0.5, width=0.5, color="lightblue")
ax.add_patch(a1)

a2 = mpatches.Arrow(x=0.3, y=0.1, dx=0.6, dy=0.5, width=0.5, color="pink")
ax.add_patch(a2)

plt.show()
```

## 矢印を描画する(細かい書式設定ができる版)

```python
# Fancy Arrow矢印を描画する
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1,1,1)

a1 = mpatches.FancyArrow(
    x=0.2, y=0.3, dx=0.4, dy=0.3, width=0.1,
    head_width=0.3, head_length=0.2, shape="right",
    color="lightblue")
ax.add_patch(a1)

a2 = mpatches.FancyArrow(
    x=0.2, y=0.3, dx=0.4, dy=0.3, width=0.1,
    head_width=0.3, head_length=0.2, shape="left",
    color="pink")
ax.add_patch(a2)

plt.show()
```

## 円形を描画する

```python
# 円形を描画する
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1,1,1)

a1 = mpatches.Circle(xy=(0.5,0.5), radius=0.4, color="lightblue")
ax.add_patch(a1)

a2 = mpatches.Circle(xy=(0.5,0.5), radius=0.2, color="white")
ax.add_patch(a2)

plt.show()
```

## 正多角形を描画する

```python
# Circle Polygonを描画
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1,1,1)

cp1 = mpatches.CirclePolygon(xy=(0.5,0.5), radius=0.4, resolution=6, color="lightblue")
ax.add_patch(cp1)

cp2 = mpatches.CirclePolygon(xy=(0.5,0.5), radius=0.2, resolution=8, color="white")
ax.add_patch(cp2)

plt.show()
```

## 楕円形を描画する

```python
## Elipseを描画
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1,1,1)

e1 = mpatches.Ellipse(xy=(0.5,0.5), width=0.8, height=0.5, angle=30, color="lightblue")
ax.add_patch(e1)

e2 = mpatches.Ellipse(xy=(0.5,0.5), width=0.5, height=0.2, angle=30, color="white")
ax.add_patch(e2)

plt.show()
```
