```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

data = [0, 1]
ax.plot(data)

# 図をインライン表示
plt.show()
```

```python
# 補足： フィギュアは、サブプロットを描画する領域
# サブプロットはグラフを描画するための領域
fig = plt.figure()

ax1 = fig.add_subplot(231) # 二行三列の１番
ax2 = fig.add_subplot(232) # 二行三列の２番
ax3 = fig.add_subplot(233) # 二行三列の３番
ax4 = fig.add_subplot(234) # 二行三列の4番
ax5 = fig.add_subplot(2, 3, 5) # 二行三列の３番 カンマ区切りやで指定する例
ax6 = fig.add_subplot(2, 3, 6) # 二行三列の4番 カンマ区切りで指定する例

for i, ax in enumerate([ax1, ax2, ax3, ax4, ax5, ax6], start=1):
    txt ='ax{0}\n23{0}'.format(i)
    ax.text(0.2, 0.4, txt, fontsize=16)

plt.show()
```

```python
# フィギュアオブジェクトの生成とサブプロットの配置を同時に行う
fig, axes = plt.subplots(2, 2)

print(type(axes), axes)
plt.show()
```
